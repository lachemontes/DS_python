You're going to work on the scatter plot with world development data: GDP per capita on the x-axis (logarithmic scale), life expectancy on the `y-axis`. The code for this plot is available in the script.<br>

As a first step, let's add axis labels and a title to the plot. You can do this with the `xlabel()`, `ylabel()` and `title()` functions, available in matplotlib.pyplot. This sub-package is already imported as plt.<br>

````python

# Sample GDP per capita (in USD) and life expectancy (in years)
countries = ["Norway", "Switzerland", "United States", "Germany", "Japan",
             "Brazil", "India", "China", "South Africa", "Nigeria", "Zimbabwe"]

gdp_cap = [75000, 71000, 63000, 52000, 42000, 15000, 7000, 12000, 6000, 2500, 1400]  # GDP per capita in USD
life_exp = [82, 83, 79, 81, 85, 75, 69, 77, 65, 55, 50]  # Life expectancy in years
````

````python
# Basic scatter plot, log scale
plt.scatter(gdp_cap, life_exp)
plt.xscale('log') 

# Strings
xlab = 'GDP per Capita [in USD]'
ylab = 'Life Expectancy [in years]'
title = 'World Development in 2007'

# Add axis labels
plt.xlabel("xlab")
plt.ylabel("ylab")


# Add title
plt.title("World Development in 2007")

# After customizing, display the plot
plt.show()
````

````python
# Sample GDP per capita (in USD) and life expectancy (in years) for different countries
countries = ["Norway", "Switzerland", "United States", "Germany", "Japan",
             "Brazil", "India", "China", "South Africa", "Nigeria", "Zimbabwe"]

gdp_cap = [75000, 71000, 63000, 52000, 42000, 15000, 7000, 12000, 6000, 2500, 1400]  # GDP per capita in USD
life_exp = [82, 83, 79, 81, 85, 75, 69, 77, 65, 55, 50]  # Life expectancy in years
pop = [5, 8, 25, 15, 12, 30, 50, 90, 60, 100, 120]  # Population (in millions)
````

````python
# Import numpy as np
import numpy as np

# Store pop as a numpy array: np_pop

np_pop = np.array(pop)
# Double np_pop
np_pop = np.array(pop) * 2

# Update: set s argument to np_pop
plt.scatter(gdp_cap, life_exp, s = np_pop)


# Previous customizations
plt.xscale('log') 
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
plt.xticks([1000, 10000, 100000],['1k', '10k', '100k'])

# Display the plot
plt.show()
````

````python
col = ['green']

# Specify c and alpha inside plt.scatter()
plt.scatter(x = gdp_cap, y = life_exp, s = np.array(pop) * 2, c = col, alpha = 0.8)

# Previous customizations
plt.xscale('log') 
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
plt.xticks([1000,10000,100000], ['1k','10k','100k'])

# Show the plot
plt.show()
````
